```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import os
from selenium.webdriver.common.by import By


# 크롬창(웹드라이버) 열기
driver = webdriver.Chrome()

#장소 접속하기
driver.get("https://www.visitjeju.net/kr/detail/view?contentsid=CONT_000000000500283")
time.sleep(1)

title = driver.find_element_by_css_selector("#content > div.cont.detail_page.detail_style > div.sub_visual_wrap > div.inner_wrap > div.sub_info_area > div.sub_info_title > h3")
review_num = driver.find_element_by_css_selector('#content > div.cont.detail_page.detail_style > div.cont_wrap.sub_visual > ul > li:nth-child(3) > p.appraisal_cnt')

#리뷰창 클릭
CSS_review="#tab3 > h2 > a"
driver.find_element(By.CSS_SELECTOR, CSS_review).click()

# 페이지 넘기기
CSS_btn_next = "#paging > a.spr_com.page-next"
def go_next_page():
    next_btn = driver.find_element_by_css_selector(CSS_btn_next)
    driver.find_element_by_class_name('spr_com.page-next').send_keys(Keys.ENTER)

    
result_df = pd.DataFrame()
num = 0
button = 71

while num < button :
    try:
        for i in range(5):
            review = driver.find_element_by_css_selector(f'#contentsReviewItem{i} > div > div.user_content > div.review.clear > p.review_txt')
            df = pd.DataFrame({'place':title.text,
                                'content': review.text}, index = [0])
            result_df = pd.concat([result_df, df], ignore_index=True)
    except NoSuchElementException:
        break
        
    page = 0 
    num += 1
    
    go_next_page()
    
result_df.head()
```
