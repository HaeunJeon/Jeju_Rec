```python
from tqdm import tqdm
from time import sleep
import os
import re
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
from hanspell import spell_checker
from konlpy.tag import Mecab 
tokenizer = Mecab()

def text_preprocessing(text): #,tokenizer
    txt1 = re.sub('[ㄱ-ㅎA-Za-z0-9]', '', text)
    txt = re.sub('[_<>\-=+,#/\?:^.@*\"※~ㆍ!』‘|\(\)\[\]`\'…》\”\“\’·]', '', txt1)
    return txt

def no_emogi(text):
    emoji_pattern = re.compile("["
            u"\U0001F600-\U0001F64F"  # emoticons
            u"\U0001F300-\U0001F5FF"  # symbols & pictographs
            u"\U0001F680-\U0001F6FF"  # transport & map symbols
            u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text) 
    
def spell_check(text):
    spelled_sent = spell_checker.check(text)
    hanspell_sent = spelled_sent.checked
    return hanspell_sent

f = open("/Users/chanseonpark/Desktop/종설/전처리_불용어/stopword.txt", 'r')
lines = f.readlines()
stopwords = []
for line in lines:
    line = line.replace('\n', '')
    stopwords.append(line)
f.close()

file_path= '/Users/chanseonpark/Desktop/종설/data/review.xlsx'
df=pd.read_excel(file_path, usecols='A:B')

with tqdm(total=len(df)) as pbar:
    for i in range(0,len(df)):
        try:
            text = no_emogi(df.loc[i]['review')
            new_text = text_preprocessing(text)
            fin_text = spell_check(new_text)
            df.loc[i]['review'] = fin_text
        except:
            df.loc[i]['review'] = "PASS"
            continue
        pbar.update(1)
        
delete = df[df['review'] == 'PASS'].index
df.drop(delete, inplace = True)
os.chdir("/Users/chanseonpark/Desktop/종설/data")
df.to_excel("final_review.xlsx")  
```
